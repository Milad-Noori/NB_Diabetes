```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import warnings
```

```python
Data = pd.read_csv("diabetes.csv")
Data
```

```python
DF = pd.DataFrame(Data.drop('Pregnancies', axis=1))
DF
```

```python
DF.describe()
```

```python
DF.isnull().sum()
```

```python
X =pd.DataFrame(DF,columns=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
Y = DF['Outcome'].values.reshape(-1 , 1)
```

```python
X_train ,X_test ,Y_train , Y_test = train_test_split(X,Y,  test_size=0.10 , random_state=0)
```

```python
Model=GaussianNB()
Model.fit(X_train ,Y_train.ravel())
Y_pred = Model.predict(X_test)
```

```python
print("Accuracy :",metrics.accuracy_score(Y_test , Y_pred))
```

```python
from sklearn.metrics import classification_report , confusion_matrix
Model.classes_
```

```python
confusion_matrix(Y , Model.predict(X))
```

```python
print(classification_report(Y , Model.predict(X)))
```

```python
P1 = Model.predict([[1,30,2,50,81,70,1]])
P1
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
k_fold=KFold(5)
print(cross_val_score(Model, X,Y.ravel(),cv=k_fold , n_jobs=1))
```
